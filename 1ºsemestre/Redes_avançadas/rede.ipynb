Criar rede direcionada

```python
import networkx as nx
G = nx.DiGraph()
G.add_nodes_from([1,2,3,4,5,6,7])
G.add_edges_from([(1,2),(2,3),(2,4),(4,5),(4,6),(5,2),(5,6),(6,7),(7,5)])
nx.draw(G,pos = nx.kamada_kawai_layout(G), arrowsize=20, node_color='white', edgecolors='black', node_size=500, with_labels = True)
```

CENTRALIDADE

```python

G =  nx.erdos_renyi_graph(5,0.5,directed=True)
nx.draw(G,pos = nx.kamada_kawai_layout(G), with_labels = True)
cent = nx.eigenvector_centrality(G)
sorted((v,f"{c:0.2f}") for v, c in cent.items())
```

```python
G =  nx.erdos_renyi_graph(5,0.5,directed=True)
cent = nx.eigenvector_centrality(G)
sorted((v,f"{c:0.2f}") for v, c in cent.items())

nx.draw(G,pos = nx.kamada_kawai_layout(G), with_labels = True,node_size = (c**2 for v,c in cent.items()))
```
