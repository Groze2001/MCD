```python
import random
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
num_plots = 100
p = 0.2
fig = plt.figure(1, figsize=(8, 8))
for node_number in range(num_plots):
    G.add_node(node_number)
    list_targets = [n for n in list(G.nodes()) if n != node_number]
    if len(list_targets) > 0 and random.uniform(0, 1) < p : G.add_edge(node_number, random.choice(list_targets))
    if node_number == (num_plots-1):
      plt.clf()
      nx.draw(G, nx.kamada_kawai_layout(G), node_size=100, with_labels=True, font_color='white')
      plt.pause(0.1)
      
```

```python
import numpy as np
import random 
import networkx as nx
n = 1000
degree_sequence = random.choices(range(1,6),k=n)
if sum(degree_sequence)%2 != 0 :
  degree_sequence[random.randrange(len(degree_sequence))]+=1
G = nx.configuration_model(degree_sequence)
self_loops=len(list(nx.selfloop_edges(G)))
dup=[]
E = list(G.edges())
for e in E: 
    if E.count(e)>1:
        dup.append(e)
duplicados = len(dup)
print(f"Número de nós: {n}")
print(f"Número de self-loops: {self_loops}")
print(f"Número de multiarestas: {duplicados}")
```
