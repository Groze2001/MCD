```python
import networkx as nx
```

```python
G = nx.Graph()
nx.add_path(G, [0, 1, 2])
nx.add_path(G, [0, 10, 2])
nx.add_path(G, [0, 5, 9])
nx.add_path(G, [5, 7, 1])
nx.draw(G, arrowsize=20, node_color='white', edgecolors='black', node_size=500, with_labels =True)
print([p for p in nx.all_shortest_paths(G, source=5, target=7)])
```
